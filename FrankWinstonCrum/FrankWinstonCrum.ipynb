```python
# Librerias a utilizar

import tabula as tb
import pandas as pd
import numpy as np
import PyPDF2
from openpyxl import load_workbook
import os
import glob
```

```python
#Variable

usuario = os.path.expanduser('~')
```

```python
def obtener_ultimo_pdf_con_nombre(nombre_buscar, ruta):
    # Crear una lista de todos los archivos PDF que contienen el nombre deseado
    archivos_pdf = glob.glob(os.path.join(ruta, f'*{nombre_buscar}*.pdf'))

    if not archivos_pdf:
        print(f"No se encontraron archivos PDF que contengan '{nombre_buscar}' en el nombre.")
        return None

    # Ordenar la lista por fecha de modificación (el último archivo será el último de la lista)
    archivos_pdf.sort(key=os.path.getmtime, reverse=True)

    # Obtener el último archivo de la lista
    ultimo_pdf = archivos_pdf[0]

    return ultimo_pdf

# Ruta donde se encuentran los archivos PDF
ruta_descargas = f'{usuario}\\Downloads\\'

# Nombre que buscamos en el archivo PDF
nombre_buscar = "METROP"

# Obtener el último archivo PDF con el nombre deseado
ultimo_pdf = obtener_ultimo_pdf_con_nombre(nombre_buscar, ruta_descargas)

if ultimo_pdf:
    print(f"El último archivo PDF con '{nombre_buscar}' en el nombre es: {ultimo_pdf}")
else:
    print("No se encontraron archivos PDF que coincidan con el nombre buscado.")
```

```python
#Rutas

file_path = ultimo_pdf
file_out = f'{usuario}\\Desktop\\Proyectos\\FrankWinstonCrum\\Entregables\\'
os.makedirs(file_out, exist_ok=True)
```

```python
#Lectura de archivo

pdf_reader = tb.read_pdf(file_path, pages="all", multiple_tables=True, stream=False, guess=True, output_format="dataframe")
len(pdf_reader) #visualizar cantidad de tablas recogidas
```

```python
type(pdf_reader) #Visualizar tipo de archivo generado
```

```python
df = pd.concat(pdf_reader) #Convertir la tabla en dataframe
df.head()
```

```python

df.drop(df[(df['Account\rNumber'] == 'Agency Totals') | (df['Account\rNumber'] == 'Agency Grand Totals')].index, inplace=True)

#Creación de columnas PH y AN

df['Policy Holder'] = df['Policy Holder'].fillna('') #Los campos en blanco los completa con vacios
df['Policy Holder_2'] = np.where(df['Policy Holder'] == '', df['Account\rNumber'].str[8:], df['Policy Holder']) #DondePH sea igual a blanco, colocame el str desde el valor 8 en adelante, sino colocame el valor original. 
df['Account\rNumber_2'] = np.where(df['Policy Holder'] == '', df['Account\rNumber'].str[:7], df['Account\rNumber']) # ""

df['Policy Number'] = df['Policy Number'].fillna('') #Los campos en blanco los completa con vacios
df['Payment\rDate'] = df['Payment\rDate'].fillna('')
df = df.loc[df['Policy Number'] != ""] #Resulta un DF con todos los varoles diferentes a blanco

df.head()
```

```python
#Limpieza de datos:

for column in df.columns:
    df[column] = df[column].str.replace('(', '-', regex=False)
    
def replace_chars_in_df(df, chars_to_replace):
    for column in df.columns:
        for char in chars_to_replace:
            df[column] = df[column].str.replace(char, '', regex=False)

chars_to_replace = ['$',')',',']
replace_chars_in_df(df, chars_to_replace)
```

```python
#Formateo de columnas:

df = df.rename(columns={'Account\rNumber_2' : 'ACCOUNT NUMBER', 'Policy Holder_2' : 'Insured Name', 'Line of\rBusiness' : 'LINE OF BUSINESS', 'Carrier' : 'CARRIER' , 'Payment\rDate' : 'Policy Eff Date', 'Commissionable\rPremium' : 'Premium', 'Commission\rRate' : 'Comm Percent', 'Commission\rAmount' : 'Comm Amount', 'Producer\rFee' : 'PRODUCER FEE', 'Total Paid' : 'TOTAL PAID'})
df["PRODUCER CODE"] = ''
df["Trans Type Code"] = 'PCH' 
df["VENDOR CODE"] = 'METROP1260'

df = df[['ACCOUNT NUMBER', 'Insured Name', 'Policy Number', 'LINE OF BUSINESS', 'CARRIER', 'Policy Eff Date', 'Premium', 'Comm Percent', 'Comm Amount', 'Trans Type Code', 'PRODUCER FEE', 'TOTAL PAID', 'PRODUCER CODE', 'VENDOR CODE']]
```

```python
#df.loc[df['ACCOUNT_NUMBER'].str[:7] == 'J205658'] #Verificar info
df.head()
```

```python
#Formateo de tipo de datos:

df['Premium'] = pd.to_numeric(df['Premium'], errors='coerce')
df['Comm Amount'] = pd.to_numeric(df['Comm Amount'], errors='coerce')
df['PRODUCER FEE'] = pd.to_numeric(df['PRODUCER FEE'], errors='coerce')
df['TOTAL PAID'] = pd.to_numeric(df['TOTAL PAID'], errors='coerce')
df['Policy Eff Date'] = pd.to_datetime(df['Policy Eff Date'], errors='coerce')
df['Policy Eff Date'] = df['Policy Eff Date'].dt.strftime('%m-%d-%Y')
```

```python
df.info()
```

```python
# Creamos una lista para almacenar los dataframes
dfs = []

# Creamos una variable para almacenar el índice del inicio de cada nuevo dataframe
start = 0

# Recorremos cada fila en el dataframe
for i in range(len(df)):
    # Si la primera columna contiene 'Account', 'ACCOUNT', o 'account'
    if df.iloc[i, 0] in ['Account', 'ACCOUNT', 'account', 'Account\rNumber']:
        # Agregamos el dataframe actual a la lista
        dfs.append(df.iloc[start:i])
        # Actualizamos el índice de inicio para el próximo dataframe
        start = i

# Agregamos el último dataframe a la lista
dfs.append(df.iloc[start:])

# # Ahora, tenemos una lista de dataframes que podemos escribir en archivos Excel
# for i, df in enumerate(dfs):
#     # Escribimos cada dataframe en un archivo Excel   
#     df.to_excel(f'dataframe{i}.xlsx')
```

```python
#Copia el resultado de la iteración
df1 = dfs[0].copy()
df2 = dfs[1].copy()
df3 = dfs[2].copy()

#Agregar valor manual
df1.loc[:, 'PRODUCER CODE'] = 'FC126001'
df2.loc[:, 'PRODUCER CODE'] = 'FC126002'
df3.loc[:, 'PRODUCER CODE'] = 'FC126003'

#Limpieza de encabezados adicionales
df2.drop(df2[(df2['CARRIER'] == 'Carrier')].index, inplace=True)
df3.drop(df3[(df3['CARRIER'] == 'Carrier')].index, inplace=True)
```

```python
#Lectura de texto fecha de reporte

archivo = open(file_path, 'rb')
pdf_reader = PyPDF2.PdfReader(archivo)
page = pdf_reader.pages[0]
text = page.extract_text()
fArchivo = text.split('\n')
fArchivo = fArchivo[1]
```

```python
#Exportar archivos excel

df1.to_excel(f'{file_out}FrankWinstonCrum-METROP126001-Statement-{fArchivo}.xlsx', sheet_name='Sheet1', index=False)
df2.to_excel(f'{file_out}FrankWinstonCrum-METROP126002-Statement-{fArchivo}.xlsx', sheet_name='Sheet1', index=False)
df3.to_excel(f'{file_out}FrankWinstonCrum-METROP126003-Statement-{fArchivo}.xlsx', sheet_name='Sheet1', index=False)
```

```python
#Auto ajustar columnas

Excel126001 = f'{file_out}FrankWinstoncrum-METROP126001-Statement-{fArchivo}.xlsx'
wb = load_workbook(Excel126001)
worksheet = wb['Sheet1']

for i, col in enumerate(df1.columns):
    column_width = max(df1[col].astype(str).map(len).max(), len(col)) + 2
    worksheet.column_dimensions[worksheet.cell(row=1, column=i + 1).column_letter].width = column_width

wb.save(Excel126001)

Excel126002 = f'{file_out}FrankWinstoncrum-METROP126002-Statement-{fArchivo}.xlsx'
wb = load_workbook(Excel126002)
worksheet = wb['Sheet1']

for i, col in enumerate(df1.columns):
    column_width = max(df1[col].astype(str).map(len).max(), len(col)) + 2
    worksheet.column_dimensions[worksheet.cell(row=1, column=i + 1).column_letter].width = column_width

wb.save(Excel126002)

Excel126003 = f'{file_out}FrankWinstoncrum-METROP126003-Statement-{fArchivo}.xlsx'
wb = load_workbook(Excel126003)
worksheet = wb['Sheet1']

for i, col in enumerate(df1.columns):
    column_width = max(df1[col].astype(str).map(len).max(), len(col)) + 2
    worksheet.column_dimensions[worksheet.cell(row=1, column=i + 1).column_letter].width = column_width

wb.save(Excel126003)
```
